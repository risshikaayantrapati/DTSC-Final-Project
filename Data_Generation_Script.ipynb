```python
import csv
from faker import Faker
import random
import numpy as np

fake = Faker()

def generate_car_rental_data(num_records=102545):
    data = []
    used_owner_ids = set()
    
    fuel_type_distribution = {'Hybrid': 0.09, 'Gasoline': 0.45, 'Electric': 0.16, 'Diesel': 0.30}
    car_type_distribution = {'Van': 0.10, 'Truck': 0.15, 'SUV': 0.30, 'Sedan': 0.30, 'Minivan': 0.15}
    
    vehicle_type_multipliers = {'Sedan': 1.0,'SUV': 1.2,'Truck': 1.5,'Van': 1.3,'Minivan': 1.4}

     # Dictionary mapping each make to a list of models
    make_model_mapping = {
        'Acura': ['RDX', 'MDX'],
        'Alfa Romeo': ['Giulia', 'Stelvio'],
        'Aston Martin': ['Vantage', 'DB11'],
        'Audi': ['A4', 'Q3', 'Q7'],
        'Bentley': ['Continental GT', 'Bentayga', 'Mulsanne'],
        'BMW': ['3 Series', 'X1', 'X5'],
        'Bugatti' : ['Veyron', 'Chiron'],
        'Buick' : ['Encore', 'Enclave'],
        'Cadillac' : ['XT5', 'Escalade'],
        'Chevrolet': ['Silverado', 'Tahoe', 'Corvette'],
        'Chrysler': ['300', 'Pacifica'],
        'Dodge': ['Journey', 'Challenger', 'Durango'],
        'Ferrari': ['812 Superfast', 'Portofino', '488 Pista'],
        'Fiat': ['500', '500C', '500X'],
        'Ford': ['Fiesta', 'Mustang', 'Ecosport'],
        'GMC' : ['Sierra', 'Terrain'],
        'Honda': ['HR-V', 'Civic', 'CR-V'],
        'Hyundai': ['Tucson', 'Santa Fe', 'Accent'],
        'Jeep': ['Cherokee', 'Grand Cherokee', 'Wrangler'],
        'Kia': ['Sorento', 'Rio', 'Sportage'],
        'Lamborghini': ['Aventador', 'Huracan'],
        'Lexus': ['IS', 'RC', 'RX'],
        'Mazda': ['CX-3', 'Mazda3', 'MX-5 Miata'],
        'Mercedes Benz': ['C-Class', 'CLA-Class', 'GLC'],
        'Nissan': ['Altima', 'Rogue', 'Titan'],
        'Porsche': ['911', 'Macan', 'Cayenne'],
        'Tesla': ['Model 3', 'Model Y'],
        'Toyota': ['Camry', 'Corolla', 'Tacoma'],
        'Volkswagen': ['Golf', 'Jetta', 'Beetle'],
        'Volvo' : ['XC90', 'XC60']
    }

    states = [
        {"name": "Alabama", "abbreviation": "AL", "cities": ["Birmingham", "Montgomery", "Mobile", "Huntsville", "Tuscaloosa"]},
        {"name": "Alaska", "abbreviation": "AK", "cities": ["Anchorage", "Fairbanks", "Juneau", "Sitka", "Ketchikan"]},
        {"name": "Arizona", "abbreviation": "AZ", "cities": ["Phoenix", "Tucson", "Mesa", "Chandler", "Scottsdale"]},
        {"name": "Arkansas", "abbreviation": "AR", "cities": ["Little Rock", "Fort Smith", "Fayetteville", "Springdale", "Jonesboro"]},
        {"name": "California", "abbreviation": "CA", "cities": ["Los Angeles", "San Francisco", "San Diego", "Sacramento", "San Jose"]},
        {"name": "Colorado", "abbreviation": "CO", "cities": ["Denver", "Colorado Springs", "Aurora", "Fort Collins", "Lakewood"]},
        {"name": "Connecticut", "abbreviation": "CT", "cities": ["Bridgeport", "New Haven", "Hartford", "Stamford", "Waterbury"]},
        {"name": "Delaware", "abbreviation": "DE", "cities": ["Wilmington", "Dover", "Newark", "Middletown", "Smyrna"]},
        {"name": "Florida", "abbreviation": "FL", "cities": ["Miami", "Orlando", "Tampa", "Jacksonville", "Fort Lauderdale"]},
        {"name": "Georgia", "abbreviation": "GA", "cities": ["Atlanta", "Savannah", "Augusta", "Columbus", "Macon"]},
        {"name": "Hawaii", "abbreviation": "HI", "cities": ["Honolulu", "Hilo", "Kailua", "Kahului", "Lihue"]},
        {"name": "Idaho", "abbreviation": "ID", "cities": ["Boise", "Meridian", "Nampa", "Idaho Falls", "Pocatello"]},
        {"name": "Illinois", "abbreviation": "IL", "cities": ["Chicago", "Aurora", "Rockford", "Joliet", "Naperville"]},
        {"name": "Indiana", "abbreviation": "IN", "cities": ["Indianapolis", "Fort Wayne", "Evansville", "South Bend", "Carmel"]},
        {"name": "Iowa", "abbreviation": "IA", "cities": ["Des Moines", "Cedar Rapids", "Davenport", "Sioux City", "Iowa City"]},
        {"name": "Kansas", "abbreviation": "KS", "cities": ["Wichita", "Overland Park", "Kansas City", "Topeka", "Olathe"]},
        {"name": "Kentucky", "abbreviation": "KY", "cities": ["Louisville", "Lexington", "Bowling Green", "Owensboro", "Covington"]},
        {"name": "Louisiana", "abbreviation": "LA", "cities": ["New Orleans", "Baton Rouge", "Shreveport", "Lafayette", "Lake Charles"]},
        {"name": "Maine", "abbreviation": "ME", "cities": ["Portland", "Lewiston", "Bangor", "South Portland", "Auburn"]},
        {"name": "Maryland", "abbreviation": "MD", "cities": ["Baltimore", "Frederick", "Rockville", "Gaithersburg", "Bowie"]},
        {"name": "Massachusetts", "abbreviation": "MA", "cities": ["Boston", "Worcester", "Springfield", "Cambridge", "Lowell"]},
        {"name": "Michigan", "abbreviation": "MI", "cities": ["Detroit", "Grand Rapids", "Warren", "Sterling Heights", "Ann Arbor"]},
        {"name": "Minnesota", "abbreviation": "MN", "cities": ["Minneapolis", "St. Paul", "Rochester", "Duluth", "Bloomington"]},
        {"name": "Mississippi", "abbreviation": "MS", "cities": ["Jackson", "Gulfport", "Southaven", "Hattiesburg", "Biloxi"]},
        {"name": "Missouri", "abbreviation": "MO", "cities": ["Kansas City", "St. Louis", "Springfield", "Independence", "Columbia"]},
        {"name": "Montana", "abbreviation": "MT", "cities": ["Billings", "Missoula", "Great Falls", "Bozeman", "Butte"]},
        {"name": "Nebraska", "abbreviation": "NE", "cities": ["Omaha", "Lincoln", "Bellevue", "Grand Island", "Kearney"]},
        {"name": "Nevada", "abbreviation": "NV", "cities": ["Las Vegas", "Henderson", "Reno", "North Las Vegas", "Sparks"]},
        {"name": "New Hampshire", "abbreviation": "NH", "cities": ["Manchester", "Nashua", "Concord", "Dover", "Rochester"]},
        {"name": "New Jersey", "abbreviation": "NJ", "cities": ["Newark", "Jersey City", "Paterson", "Elizabeth", "Edison"]},
        {"name": "New Mexico", "abbreviation": "NM", "cities": ["Albuquerque", "Las Cruces", "Rio Rancho", "Santa Fe", "Roswell"]},
        {"name": "New York", "abbreviation": "NY", "cities": ["New York City", "Buffalo", "Rochester", "Yonkers", "Syracuse"]},
        {"name": "North Carolina", "abbreviation": "NC", "cities": ["Charlotte", "Raleigh", "Greensboro", "Durham", "Winston-Salem"]},
        {"name": "North Dakota", "abbreviation": "ND", "cities": ["Fargo", "Bismarck", "Grand Forks", "Minot", "West Fargo"]},
        {"name": "Ohio", "abbreviation": "OH", "cities": ["Columbus", "Cleveland", "Cincinnati", "Toledo", "Akron"]},
        {"name": "Oklahoma", "abbreviation": "OK", "cities": ["Oklahoma City", "Tulsa", "Norman", "Broken Arrow", "Lawton"]},
        {"name": "Oregon", "abbreviation": "OR", "cities": ["Portland", "Eugene", "Salem", "Gresham", "Hillsboro"]},
        {"name": "Pennsylvania", "abbreviation": "PA", "cities": ["Philadelphia", "Pittsburgh", "Allentown", "Erie", "Reading"]},
        {"name": "Rhode Island", "abbreviation": "RI", "cities": ["Providence", "Warwick", "Cranston", "Pawtucket", "East Providence"]},
        {"name": "South Carolina", "abbreviation": "SC", "cities": ["Columbia", "Charleston", "North Charleston", "Mount Pleasant", "Rock Hill"]},
        {"name": "South Dakota", "abbreviation": "SD", "cities": ["Sioux Falls", "Rapid City", "Aberdeen", "Brookings", "Watertown"]},
        {"name": "Tennessee", "abbreviation": "TN", "cities": ["Nashville", "Memphis", "Knoxville", "Chattanooga", "Clarksville"]},
        {"name": "Texas", "abbreviation": "TX", "cities": ["Houston", "San Antonio", "Dallas", "Austin", "Fort Worth"]},
        {"name": "Utah", "abbreviation": "UT", "cities": ["Salt Lake City", "West Valley City", "Provo", "West Jordan", "Orem"]},
        {"name": "Vermont", "abbreviation": "VT", "cities": ["Burlington", "Essex", "South Burlington", "Colchester", "Rutland"]},
        {"name": "Virginia", "abbreviation": "VA", "cities": ["Virginia Beach", "Norfolk", "Chesapeake", "Richmond", "Newport News"]},
        {"name": "Washington", "abbreviation": "WA", "cities": ["Seattle", "Spokane", "Tacoma", "Vancouver", "Bellevue"]},
        {"name": "West Virginia", "abbreviation": "WV", "cities": ["Charleston", "Huntington", "Parkersburg", "Wheeling", "Morgantown"]},
        {"name": "Wisconsin", "abbreviation": "WI", "cities": ["Milwaukee", "Madison", "Green Bay", "Kenosha", "Racine"]},
        {"name": "Wyoming", "abbreviation": "WY", "cities": ["Cheyenne", "Casper", "Laramie", "Gillette", "Rock Springs"]},
    ]
    major_cities_distribution = {'Los Angeles': 0.15, 'New York City': 0.12, 'Chicago': 0.1, 'Houston': 0.08, 'Miami': 0.07}
    repetitive_cities_with_high_probs = [city for city, _ in major_cities_distribution.items()]
    #repetitive_cities = [city for state in states for city in state["cities"]]

    for _ in range(num_records):
        #city = random.choice(repetitive_cities) if random.random() < 0.2 else fake.city()  # 20% chance to choose a repetitive city
        city = random.choice(repetitive_cities_with_high_probs) if random.random() < 0.3 else fake.city()  # 30% chance to choose a major city
        state = get_state_for_city(city, states)

        make = fake.random_element(elements=list(make_model_mapping.keys()))
        model = fake.random_element(elements=make_model_mapping[make])
        
        
        vehicle_data = {
            'make': make,
            'model': model,
            'year': random.randint(2000, 2023),
        }
        
        vehicle_data['type'] = generate_custom_car_type(car_type_distribution)  # Add the 'type' key based on car type
        
        
        type_multiplier = vehicle_type_multipliers.get(vehicle_data['type'], 1.0)
        city_multiplier = 1.5 if city in major_cities_distribution else 1.0  # Adjust the multiplier based on whether it's a major city or not
        
        # Adjust the daily rate based on vehicle type and city
        base_daily_rate = round(random.uniform(30.0, 1500.0), 2)
        adjusted_daily_rate = base_daily_rate * type_multiplier * city_multiplier
        
        rating_prob = random.random()
        if rating_prob < 0.5:
            rating = round(random.uniform(3.0, 5.0), 1)
        elif rating_prob < 0.7:
            rating = 4.4
        elif rating_prob < 0.9:
            rating = 4.8
        else:
            rating = round(random.choice([3.0, 3.3, 3.5, 3.6, 4.0, 4.1, 4.5, 4.7, 4.8, 5.0]), 1)
        
        trips_multiplier = 1.5 if city in major_cities_distribution else 1.0  # Adjust the multiplier based on whether it's a major city or not
        base_trips = random.randint(50, 200)
        renter_trips_taken = int(base_trips * trips_multiplier)
        
        record = {
            'fuelType': generate_custom_fuel_type(fuel_type_distribution),
            'rating': rating,
            'renterTripsTaken': renter_trips_taken,
            'reviewCount': random.randint(0, 350),
            'city': city,
            'Country': 'US',
            'state': state,
            'latitude': round(random.uniform(24.396308, 49.384358), 6),
            'longitude': round(random.uniform(-125.000000, -66.934570), 6),
            'owner_id': generate_unique_owner_id(used_owner_ids),
            'type': generate_custom_car_type(car_type_distribution),
            'daily_rate': adjusted_daily_rate,
            **flatten_vehicle(vehicle_data),
            'airport_city': city,
        }
        data.append(record)

    return data

def generate_custom_fuel_type(fuel_type_distribution):
    """Generate a fuel type based on the specified distribution."""
    fuel_types = list(fuel_type_distribution.keys())
    probabilities = list(fuel_type_distribution.values())

    # Normalize probabilities to ensure they sum to 1
    probabilities /= np.sum(probabilities)

    # Sample from a multinomial distribution
    selected_fuel_type = np.random.choice(fuel_types, p=probabilities)

    return selected_fuel_type

def generate_unique_owner_id(used_ids):
    """Generate a unique 5 to 6-digit owner_id."""
    while True:
        owner_id = random.randint(10000, 999999)
        if owner_id not in used_ids:
            used_ids.add(owner_id)
            return owner_id

def flatten_vehicle(vehicle):
    """Flatten the 'vehicle' dictionary."""
    flattened_vehicle = {
        'make': vehicle['make'],
        'model': vehicle['model'],
        'year': vehicle['year'],
    }
    return flattened_vehicle

def get_state_for_city(city, states):
    """Get the state abbreviation for a given city (case-insensitive)."""
    for state_info in states:
        if city.lower() in [c.lower() for c in state_info["cities"]]:
            return state_info["abbreviation"]
    return random.choice(states)["abbreviation"]

def generate_custom_car_type(car_type_distribution):
    """Generate a car type based on the specified distribution."""
    rand_val = random.random()
    cumulative_prob = 0.0
    for car_type, prob in car_type_distribution.items():
        cumulative_prob += prob
        if rand_val <= cumulative_prob:
            return car_type
        
def write_to_csv(data, filename='car_data.csv'):
    with open(filename, 'w', newline='') as csvfile:
        fieldnames = data[0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for record in data:
            writer.writerow(record)

if __name__ == "__main__":
    num_records = 102545
    car_data = generate_car_rental_data(num_records)

    # Write the data to a CSV file
    write_to_csv(car_data)
```
